```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3],[4,5,1])
plt.show()
```

#TITLE

```python
x=[1,2,3]
y=[4,5,1]
plt.plot(x,y)
plt.title("1d-graph")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.show()
```

#ADD STYLE

```python
from matplotlib import style
style.use("ggplot")

x1=[1,2,3]
y1=[4,5,2]

x2=[4,5,2]
y2=[1,2,3]

plt.plot(x1,y1,'g',label='True line',linewidth=5)  # 'g' To mention the color of the line
plt.plot(x2,y2,'b',label='False line',linewidth=5)  # 'b' To mention the color of the line

plt.title("1d-graph")
plt.ylabel("y-axis")
plt.xlabel("x-axis")

plt.legend() # To mention line names on the top of the graph
plt.grid(True,color='k')  # To draw black line in the background of the graph
plt.show()
```
