```python
import matplotlib.pyplot as plt
```

```python
slices=[3,10,10,1]
activities=['eating','working','sleeping','playing']
colors=['r','y','b','g']
```

```python
plt.pie(slices,labels=activities,colors=colors,startangle=90,shadow=True,explode=(0,0.1,0,0),autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```
