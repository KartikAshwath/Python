```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
eating=[2,4,6,1,5,9,4]
working=[3,1,0,5,2,8,1]
sleeping=[5,8,6,9,5,8,2]
playing=[1,0,2,2,0,1,1]
```

```python
plt.plot([],[],color='r',label='eating',linewidth=5)
plt.plot([],[],color='k',label='working',linewidth=5)
plt.plot([],[],color='b',label='sleeping',linewidth=5)
plt.plot([],[],color='g',label='playing',linewidth=5)

plt.stackplot(days,eating,working,sleeping,playing, colors=['r','k','b','g'])

plt.xlabel('activities')
plt.ylabel('days')
plt.title('Stack/Area')
plt.legend()
plt.show()
```
